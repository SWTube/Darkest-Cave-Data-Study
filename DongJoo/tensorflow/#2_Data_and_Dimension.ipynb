# 차원
## 표의 열(칼럼) VS 포함 관계
### 1. 칼럼 <br>
### = 변수 <br>
★ 대응관계: 표(데이터) ↔ 공간 <br>
* 열의 개수 = 변수의 개수 = 공간의 차원 수(N) <br>
* 관측치 = N차원 공간 위 한 점 <br>
데이터 間 거리 측정 & 정밀 비교 가능 <br>
*** 
### 2. 포함 관계 <br>
### = 배열의 깊이 <br>
 - n차원 공간의 관측치(all 변수's 각각의 값) 한개가 하나의 배열에 대입된 형태(x1): `1차원 배열`<br>
ex) x1.shape = (n, ) <br>
→ 숫자 1개. 1차원  |  n: 표의 변수 개수(차원) <br>
 - 1차원 배열을 요소로 갖는 배열(y1): `2차원 배열` <br>
ex) y1.shape = (n, m) <br>
→ 숫자 2개. 2차원  |  n: 1차원 배열 개수<행>, m: 표의 변수 개수(차원)<열> <br>
*** 
### ∴ x차원 배열을 요소로 갖는 배열(z): `(x+1)차원 배열`
