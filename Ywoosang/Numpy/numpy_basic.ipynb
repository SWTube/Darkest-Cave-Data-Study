```python
# numpy    
import numpy as np 

array1 = np.array([1,2,3,4])
# 2*2    
array2 = array1.reshape((2,2)) 
print(array2) 
```

```python
# numpy    
import numpy as np
# 0~4     2   1*4  
array1 = np.arange(4).reshape(1,4)
# 0~7     2   2*4 
array2 = np.arange(8).reshape(2,4) 
print(array1)
print(array2)

#    
# axis = 0 (x   )
# axis =0          
array3 = np.concatenate([array1,array2],axis = 0) 
print(array3)
```

```python
# numpy  
import numpy as np

# reshape  2  2*4   
array = np.arange(8).reshape(2,4)
# ,  split        0,1  left, 2,3  right 
# axis = 1 (  )
left, right = np.split(array,[2],axis=1)
#  
print(left.shape)
print(left.shape)

#  array
print(' \n',array)

# left,right  
print(left)
print(right) 

#   
# axis = 0  x (row)  
array = np.arange(8).reshape(2,4)
#  
top,bottom = np.split(array,[1],axis=0) 
print(top)
print(bottom)
```
