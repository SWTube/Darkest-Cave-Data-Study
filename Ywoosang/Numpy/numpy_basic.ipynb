```python
import numpy as np 
# 리스트 자료형보다 더 많은 기능을 제공하기 위한 라이브러리

# 리스트를 만듬 
list_data = [1,2,3]
# 특정한 리스트 데이터가 안에 들어가 있을 때 넘파이 자료형으로 바꿔줌
array = np.array(list_data) 

# 넘파이 데이터기 떄문에 size, dtype 과 같은 다양한 함수 제공
print(array) 
print(array.size)
print(array.dtype)
print(array[2])
```

```python
# 넘파이 배열의 초기화
# 행렬 곱 등 기본적인 연산을 제공해주는 기능 

# 0 부터  3 까지의 배열 만들기 
array1 = np.arange(4) 
print(array1) 

# 특정한 데이터를 만드는데 각 데이터가 0으로 초기화 되어 있으면 좋겠는 경우
# dtype 파라미터를 받으면 그 타입에 해당하는 데이터를 만들어 줌
# 4,4 크기의 2차원 배열을 만들고 그 모든값들이 0 으로 초기화 되어 있으면 float 타입 
array2 = np.zeros((4,4),dtype=float)
print(array2)

# 1 로 초기화
array3 = np.ones((3,3),dtype=str)
print(array3)
```

```python
import numpy as np

# 0 부터 9까지 랜덤하게 초기화된 배열 만들기
# 3,3크기의 배열 데이터가 0 부터 9 사이 임의로 채워짐
array4 = np.random.randint(0,10,(3,3)) 
print(array4)

# 특정한 통계적 데이터 만들기 가능
# 평균이 0 이고, 표준편차가 1인 표준 정규를 띄는 배열 
array5 = np.random.normal(0,1,(3,3))
print(array5) 
```

```python
# numpy 배열 합치기
import numpy as np

array1 = np.array([1,2,3,4])
array2 = np.array([5,6,7,8])
array3 = np.concatenate([array1,array2])
# 크기 
print(array3.shape)
# 형태
print(array3) 
```

```python
# numpy 배열 형태 바꾸기 
import numpy as np 

array1 = np.array([1,2,3,4])
# 2*2 형태로 바꿀 때 
array2 = array1.reshape((2,2)) 
print(array2) 
```

```python
# numpy 배열 세로 축으로 합치기
import numpy as np
# 0~4 까지 데이터를 만든 다음 2차원 배열 형태로 1*4  
array1 = np.arange(4).reshape(1,4)
# 0~7 까지 데이터를 만든 다음 2차원 배열 형태로 2*4 
array2 = np.arange(8).reshape(2,4) 
print(array1)
print(array2)

# 두 배열을 세로축으로 합침
# axis = 0 (x 축 을 합침)
# axis =0 으로 하면 첫 인수 행의 아래쪽에 두번째 인수 데이터가 들어감
array3 = np.concatenate([array1,array2],axis = 0) 
print(array3)
```

```python
# numpy 배열 나누기
import numpy as np

# reshape 로 2차원 배열 2*4 형태로  
array = np.arange(8).reshape(2,4)
# 왼쪽, 오른쪽으로 split 하는데 가로 축 두 번쨰 인덱스를 기준으로 0,1 인덱스는 left, 2,3 인덱스는 right 
# axis = 1 ( 기준 인덱스)
left, right = np.split(array,[2],axis=1)
# 형태 보기
print(left.shape)
print(left.shape)

# 원래 array
print('원본 \n',array)

# left,right 나눠서 
print(left)
print(right) 

# 위 아래 나누기
# axis = 0  x 축(row) 기준으로 
array = np.arange(8).reshape(2,4)
# 첫번째 
top,bottom = np.split(array,[1],axis=0) 
print(top)
print(bottom)
```
