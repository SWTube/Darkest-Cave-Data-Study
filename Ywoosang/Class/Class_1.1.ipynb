## 기계학습

현재 : 지식기반 -> 기계학습(ML)
  
인공지능(AI) : 대체적으로 학습기반(ML), 지식기반 둘다 포함한 개념
``` 
AI > ML > DL
``` 
기계학습 (ML)
- 데이터 중심 접근 방식  
- 신경망방식과 비 신경방 방식으로 나뉨  
  
신경망 (Neural Network) 
- 기계학습(ML)에 속하는 한가지 방법
- Deep/ Swallow 신경망으로 나뉨
  
딥러닝 (DL)
- Deep 한 신경망 기반의 ML

## Supervised Learning(지도 학습) 

주어진 input 에서 output 을 예측  
- input 에 따른 output y 이 뭐가 나올지 아는 상황 (already know what our output should look like) 
- 따라서 정확한 예측 결과에 기반해 직접적인 피드백을 얻을 수 있음  
      
키워드    
- Regression : output y 가 연속적인 경우 (ex 테스트집합에 따라 output 값이 연속적으로 나옴)
 * 특징에 따라 직선형 함수 관계를 찾음 
- Classification : 분류 가능한 문제 (ex 머리카락 길이와 키로 남학생,여학생인지 분류 가능)  
 * 결정 경계를 찾음 (decision boundary)

## Unsupervised Learning(비지도 학습) 

데이터에서 숨겨진 구조를 찾아냄
- 결과값이 어떤식으로 나올지 알 수 없는 상황
- 따라서 결과예측을 기반으로 피드백을 받을 수 없음
- 데이터 안 변수간의 관계에서 군집화(clustering) 을 통해 구조를 이끌어냄  
키워드  
- clustering  
- dimension reduction
